```python
import os
import csv
import pandas as pd
from skimage import io

i=0
img_file = []
img_size = []
img_n_channels = []
img_shape = []
z_depth = []

for subdir, dirs, files in os.walk('G:/2016/Summer/LungMAP/Image_data/'):
    for file in files:
        #print os.path.join(subdir, file)
        filepath = subdir + os.sep + file
       
        if filepath.endswith(".tif"):
            i=i+1
            img_file.append(file)
            s = os.stat(filepath)
            img_size.append(s.st_size)
            im = io.imread(filepath)
            img_n_channels.append(im.shape[2])
            h, w = im.shape[:2]
            img_shape.append((h, w))
            img_type = im.dtype
            if (img_type == "uint16"):
                z_depth.append(16*im.shape[2])
            elif (img_type == "uint8"):
                z_depth.append(8*im.shape[2])
            print(filepath)
    
    output = pd.DataFrame({ 'img_file': img_file,
                            'img_size': img_size,
                            'img_n_channels': img_n_channels,
                            'img_shape': img_shape,
                            'img_z_depth': z_depth
          
        })
    output.to_csv("07_metadata_image_details_1-1000.csv",index=False)
print("write to csv complete")
```
